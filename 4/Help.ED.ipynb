```python
# Importing Pandas and Numpy libraries
import pandas as pd
import numpy as np

# Reading data from file and marking "Year" as an integer rather than a string
superheroes = pd.read_csv("superhero_data.csv", parse_dates = ["FIRST_YEAR"])

# Printing the data for a quick look
superheroes.head(9)
```

```python
print("\nDo villains really have yellow eyes or red hair? lets find out!\n\n---------------------------------------------------------------\n")

# Setting up a matrix for identifing characters and comparing the ones with/without red hair and yellow eyes
characters = pd.crosstab(index = superheroes["EYES"] == "Yellow Eyes", columns = superheroes["HAIR"] == "Red Hair")

characters.columns = ["Red Hair", "Not Red"]
characters.index = ["Yellow Eyes", "Not Yellow"]

print(characters)
print("\nThis is the general overview of all the characters in our data\n\n---------------------------------------------------------------\n")

# Pulling, organizing and displaying data based on morality and eye color
villain_eyes = pd.crosstab(index = superheroes["ALIGN"] == "Bad Characters",
                           columns = superheroes["EYES"] == "Yellow Eyes").apply(lambda r: r/r.sum())

villain_eyes.columns = ["Yellow Eyes", "Not Yellow"]
villain_eyes.index = ["Bad", "Not Bad"]

print(villain_eyes)

# Calculating and displaying the percentage of villains with yellow eyes
percent_eyes = round((0.591065 * 100))
print("\nPercentage of villains with yellow eyes = ~", percent_eyes, "%")

# Calculating and displaying the percentage of villains without yellow eyes
percent_eyes = round((0.478134 * 100))
print("Percentage of villains without yellow eyes = ~", percent_eyes, "%")

print("\n---------------------------------------------------------------\n")
# Pulling, organizing and displaying data based on morality and hair color
villain_hair = pd.crosstab(index = superheroes["ALIGN"] == "Bad Characters",
                           columns = superheroes["HAIR"] == "Red Hair").apply(lambda r: r/r.sum())

villain_hair.columns = ["Red Hair", "Not Red"]
villain_hair.index = ["Bad", "Not Bad"]

print(villain_hair)

# Calculating and displaying the percentage of villains with red hair
percent_hair = round((0.5848 * 100))
print("\nPercentage of villains with red hair = ~", percent_hair, "%")

# Calculating and displaying the percentage of villains without red hair
percent_hair = round((0.683776 * 100))
print("Percentage of villains without red hair = ~", percent_hair, "%\n\n---------------------------------------------------------------")
```
