## **If you had to make a general guess, what would you say is the most dominant *Hair-Color* and *Eye-Color* that *Bad-Guys* represent in the Marvel & DC universes? How about the *Good-Guys*?**
#### *- Let's dive into the data for a closer look!*

```python
# Importing Pandas and Numpy libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# The following allows us to display our graphs in line
%matplotlib inline

# Removing warnings
import warnings
warnings.filterwarnings('ignore')

# Reading data from file and marking "Year" as an integer rather than a string
superheroes = pd.read_csv("superheroData.csv", parse_dates = ["FIRST_YEAR"])

# Printing the data for a quick look
superheroes.head()
```

#### *When dealing with a large amount of information, we are bound to have over-categorization and unfiltered noise. Therefore it is essential to decompose the gathered data, abstract the answers we are trying to find, and build a roadmap to reach our desired report. Consequently, we aggregated the various colors in the data and grouped them based on a White, Black, Grey, and ROYGBIV scale.*

```python
# Replacing repeated variables and grouping unhelpful info
# Eyes
superheroes = superheroes.replace("ND", "Unknown")
superheroes = superheroes.replace("No Eyes", "Unknown")
superheroes = superheroes.replace("One Eye", "Unknown")
superheroes = superheroes.replace("Variable", "Unknown")
superheroes = superheroes.replace("Compound Eyes", "Unknown")
superheroes = superheroes.replace("Photocellular Eyes", "Unknown")
superheroes = superheroes.replace("Variable Eyes", "Unknown")
superheroes = superheroes.replace("Multiple Eyes", "Unknown")
superheroes = superheroes.replace("Multiple Eyes", "Unknown")
superheroes = superheroes.replace("Hazel Eyes; White Eyes", "Unknown")

superheroes = superheroes.replace("Brown Eyes formerly Blue Eyes", "Brown Eyes")
superheroes = superheroes.replace("Red Eyes formerly Blue Eyes", "Red Eyes")
superheroes = superheroes.replace("Brown Eyes formerly Amber formerly Green", "Brown Eyes")
superheroes = superheroes.replace("Blue Eyes formerly Green", "Blue Eyes")
superheroes = superheroes.replace("Yellow Eyeballs", "Yellow Eyes")
superheroes = superheroes.replace("Black Eyeballs", "Black Eyes")
superheroes = superheroes.replace("Magenta Eyes", "Red Eyes")
superheroes = superheroes.replace("Auburn Eyes", "Brown Eyes")
superheroes = superheroes.replace("Silver Eyes", "Grey Eyes")
superheroes = superheroes.replace("Amber Eyes", "Yellow Eyes")
superheroes = superheroes.replace("Gold Eyes", "Yellow Eyes")
superheroes = superheroes.replace("Pink Eyes", "Red Eyes")
superheroes = superheroes.replace("Hazel Eyes", "Brown Eyes")
superheroes = superheroes.replace("Purple Eyes", "Violet Eyes")
superheroes = superheroes.replace("Brown Eyes", "Orange Eyes")

# Hair
superheroes = superheroes.replace("No Hair", "Unknown")
superheroes = superheroes.replace("Bald", "Unknown")
superheroes = superheroes.replace("Bald formerly Brown Hair and sometimes Blond Hair", "Unknown")
superheroes = superheroes.replace("Dyed Hair", "Unknown")
superheroes = superheroes.replace("Brown or Green Hair", "Unknown")
superheroes = superheroes.replace("Variable Hair", "Unknown")

superheroes = superheroes.replace("Strawberry Blond Hair", "Blond Hair")
superheroes = superheroes.replace("Yellow Hair", "Blond Hair")
superheroes = superheroes.replace("Gold Hair", "Blond Hair")
superheroes = superheroes.replace("Platinum Blond Hair", "Blond Hair")
superheroes = superheroes.replace("Reddish Blond Hair", "Blond Hair")
superheroes = superheroes.replace("Strawberry Blond", "Blond Hair")
superheroes = superheroes.replace("No Hair formerly Blond Hair", "Blond Hair")
superheroes = superheroes.replace("Light Brown Hair", "Brown Hair")
superheroes = superheroes.replace("Reddish Brown Hair", "Brown Hair")
superheroes = superheroes.replace("Orange-brown Hair", "Brown Hair")
superheroes = superheroes.replace("Brown Hair dyed Black", "Brown Hair")
superheroes = superheroes.replace("Brown Hair; No Hair", "Brown Hair")
superheroes = superheroes.replace("Red Hair formerly dyed Black", "Red Hair")
superheroes = superheroes.replace("Black Hair (shaved bald)", "Black Hair")
superheroes = superheroes.replace("Black Hair with White streaks", "Black Hair")
superheroes = superheroes.replace("Gray Hair dyed Black", "Grey Hair")
superheroes = superheroes.replace("Black Hair shaved bald", "Black Hair")
superheroes = superheroes.replace("Black Hair (shaved)", "Black Hair")
superheroes = superheroes.replace("Gray Hair", "Grey Hair")
superheroes = superheroes.replace("Blond Hair", "Yellow Hair")
superheroes = superheroes.replace("Auburn Hair", "Brown Hair")
superheroes = superheroes.replace("Purple Hair", "Violet Hair")
superheroes = superheroes.replace("Pink Hair", "Red Hair")
superheroes = superheroes.replace("Silver Hair", "Grey Hair")
superheroes = superheroes.replace("Magenta Hair", "Red Hair")
superheroes = superheroes.replace("Bronze Hair", "Brown Hair")
superheroes = superheroes.replace("Brown Hair", "Orange Hair")

# Align
superheroes = superheroes.replace("Neutral Characters", "Unknown")
superheroes = superheroes.replace("Reformed Criminals", "Good Characters")

# Droping all the Unknowns
superheroes = superheroes.drop(superheroes[superheroes.EYES == "Unknown"].index)
superheroes = superheroes.drop(superheroes[superheroes.HAIR == "Unknown"].index)
superheroes = superheroes.drop(superheroes[superheroes.ALIGN == "Unknown"].index)
```

#### *Once our data has been scrubbed and cleaned for analysis, we will need to redefine its categorization and labeling for future use.*
#### *- Let's see what our data looks like after preparation.*

```python
# Here we count all the unique values in the 'Eyes' column in order to get the breakdown of colors
eyes_count = superheroes["EYES"].value_counts()
print("-------------------------")
print("       Eye Colors\n-------------------------\n")
print(eyes_count)

# Here we count all the unique values in the 'Hair' column in order to get the breakdown of colors
hair_count = superheroes["HAIR"].value_counts()
print("\n-------------------------")
print("       Hair Colors\n-------------------------\n")
print(hair_count)

# Here we count all the unique values in the 'Align' column in order to get the breakdown of colors
align_count = superheroes["ALIGN"].value_counts()
print("\n--------------------------")
print("   Character Alignment\n--------------------------\n")
print(align_count)
```

#### *Next, we will need to visualize our findings by cross-correlating the three factors we are looking at; Hair Color, Eye Color, and Alignment.*

```python
# Setting values for visualization in a pie chart
align_counts = superheroes["ALIGN"].value_counts()

# Here we are able to control the visual size of our figures
plt.rcParams["figure.figsize"] = (6, 6)
plt.rcParams['text.color'] = 'r'

# Using a pie chart, we aim to take a better look at the spread
# autopct, automatically generates percent values, with each piece of the pie having a minimum representation
align_counts.plot(kind = "pie", autopct = "%1.2f%%", subplots = True, fontsize = 12)
plt.title("Alignment of Characters", color = "blue", fontsize = 22)

# Making the plot have even axes so the pie chart is perfectly circular
plt.axis("equal")

# Getting rid of the y-axis label - http://stackoverflow.com/questions/34094596/python-matplotlib-pyplot-pie-charts-how-to-remove-the-label-on-the-left-side
# These modifications also help make the bar chart look nicer and more readable
plt.axes().set_ylabel('')

# Displaying the plot
plt.show()
```

### **We can see the good guys outnumber the bad ones!**

```python
# Here we are able to control the visual size of our figures
plt.rcParams["figure.figsize"] = (18, 18)
plt.rcParams['text.color'] = 'k'

# Setting up the needed data frame
colors_count = pd.crosstab(superheroes['EYES'], superheroes['HAIR'])

# Using a bar chart, we aim to take a better look at the eye/hair color distribution
colors_count.plot(kind= 'bar', stacked = True, width = 0.75)
plt.title('Number of Characters based on Eye & Hair Color', fontsize = 18)
plt.xlabel('')
plt.ylabel("Number of Characters", fontsize = 22)
plt.xticks(rotation = 20, fontsize = 12)
plt.yticks(fontsize = 12)

# Displaying the plot
plt.show()
```

### **The three most dominant combinations seem to be *Yellow-Hair and Blue-Eyes* in the first place and *Yellow-Hair and Orange-Eyes* in second place.**

```python
# Setting up the data frame we need for our plot
align_eyes = pd.crosstab(superheroes["ALIGN"], superheroes["HAIR"])
align_eyes.plot(kind = 'bar')

# Creating our area plot for the given variables
plt.title("Number of Characters based on Hair Color & Alignment", fontsize = 22)
plt.xticks(rotation = 0, fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel("Number of Characters", fontsize = 12)
plt.ylabel('')

# Displaying the plot
plt.show()
```

## **Below are the two top trends:**
### 1. *Black-Hair* is the first dominant color for *Bad-Guys*, but there are more *Good-Guys* with *Black-Hair* than bad.
### 2. *Orange-Hair* is the second dominant color for *Bad-Guys*, but there are more *Good-Guys* with *Orange-Hair* than bad.
